#  Import necessary libraries (matplotlib and seaborn). Load the csv file of dataset into dataframe in Python program.

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/us-weather-history/KCLT.csv')
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

#  Plot	min and max for actual_mean_temp, actual_min_temp and actual_max_temp using bar graph. Use blue for min and red for max.

```python
x=['actual_mean_temp','actual_max_temp','actual_min_temp']
y1 = []
y2 = []
def values(d):
  y1.append(d.min())
  y2.append(d.max())
values(df['actual_max_temp'])
values(df['actual_mean_temp'])
values(df['actual_min_temp'])
plt.bar(x, y1, color='blue')
plt.bar(x, y2,bottom=y1 ,color='red')
plt.show()
```

# Plot	two separate histograms for actual_mean_temp for months of Aug 2014 and March 2015.

```python
#getting indices for aug 2014

i = (df.loc[df.date=='2014-8-1']).index
i=int(i.tolist()[0])
print(i)

j = (df.loc[df.date=='2014-8-31']).index
j=int(j.tolist()[0])
print(j)
```

```python
#putting the data of aug 2014 into a dataframe
aug = df['actual_mean_temp'].iloc[i:j+1]
aug
```

```python
#getting data for march 2051
k = (df.loc[df.date=='2015-3-1']).index
k=int(k.tolist()[0])
print(k)

l = (df.loc[df.date=='2015-3-31']).index
l = int(l.tolist()[0])
print(l)
```

```python
#putting the data of march 2015 into a dataframe
mar = df['actual_mean_temp'].iloc[k:l+1]
mar
```

```python
# histogram for aug 2014
plt.hist(aug, edgecolor='black') 
plt.title("Actual_mean_temp for August 2014") 
plt.show()
```

```python
# histogram for march 2015
plt.hist(mar, edgecolor='black') 
plt.title("Actual_mean_temp for March 2015") 
plt.show()
```

# Find the correlation between the following pairs of attributes: actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. 

```python
df['actual_mean_temp'].corr(df['actual_min_temp'])
```

```python
df['actual_mean_temp'].corr(df['actual_max_temp'])
```

**Observation:** both the correlations are more than 0.5 thus, are high level correlations.
